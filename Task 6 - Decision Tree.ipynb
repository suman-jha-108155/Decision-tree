# Importing Necessary Packages

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn import tree
```

```python
iris = load_iris()
iris
```

```python
#these are our features
iris.feature_names
```

```python
#these are our classifications or labels
iris.target_names
```

```python
#separating the dataframe into training and testing dataset

test_indx = [0,50,100] # for future testing

train_target = np.delete(iris.target, test_indx)
train_feature = np.delete(iris.data, test_indx, axis=0)

#for testing
test_target = iris.target[test_indx]
test_feature = iris.data[test_indx]
```

```python
#now lets have our classifier

clr = tree.DecisionTreeClassifier()
```

```python
#now lets fit the data
clr.fit(train_feature, train_target)
```

```python
predicted = clr.predict(test_feature)
```

now lets compare our prediction to original results

```python
print(test_target)
print(predicted)
```

# Visualizing the Decision Tree
